```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load the pickle file
file_path = '/Users/baileyng/MIND_models/QuantNets/Experiments_FC/run_GF_regression_lr_0.0001_scheduler_step/sgcn/test_mse.pk'

with open(file_path, 'rb') as f:
    train_mse = pickle.load(f)

# Display the contents
print("Type:", type(train_mse))
print("Shape:", np.array(train_mse).shape if hasattr(train_mse, '__len__') else "Scalar")
print("Contents:", train_mse)

# If it's a list/array of MSE values over epochs, plot them
if isinstance(train_mse, (list, np.ndarray)) and len(train_mse) > 1:
    plt.figure(figsize=(10, 6))
    plt.plot(train_mse)
    plt.title('Training MSE over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.grid(True)
    plt.show()

# Print some statistics if it's an array
if isinstance(train_mse, (list, np.ndarray)):
    train_mse_array = np.array(train_mse)
    print(f"\nStatistics:")
    print(f"Min MSE: {np.min(train_mse_array):.6f}")
    print(f"Max MSE: {np.max(train_mse_array):.6f}")
    print(f"Final MSE: {train_mse_array[-1]:.6f}")
    print(f"Mean MSE: {np.mean(train_mse_array):.6f}")
```
