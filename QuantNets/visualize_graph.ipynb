# Set Up

```python
import torch
from torch_geometric.utils import to_networkx
import matplotlib.pyplot as plt
import networkx as nx
```

# Visualize FC

```python
# Load the graph
data = torch.load('/Users/baileyng/MIND_data/DSST_FC_graph/1000184_25752_2_0_fc_graph.pt', weights_only=False)

# Convert PyTorch Geometric graph to NetworkX
G = to_networkx(data, to_undirected=True)

# Create visualization
plt.figure(figsize=(12, 8))

# Draw the graph
pos = nx.spring_layout(G, k=1, iterations=50)  # Position nodes using spring layout
nx.draw(G, pos, 
        node_color='lightblue', 
        node_size=300, 
        with_labels=True, 
        font_size=8, 
        font_weight='bold',
        edge_color='gray',
        alpha=0.7)

plt.title("Functional Connectivity Graph Visualization")
plt.axis('off')
plt.tight_layout()
plt.show()

# Print some basic graph statistics
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Graph density: {nx.density(G):.4f}")
```
